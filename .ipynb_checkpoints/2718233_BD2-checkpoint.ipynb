Step 2: Data Understanding and Quality Check

Before cleaning and merging, the datasets are explored to assess the structure and quality.

In this step:

- Column names are reviewed for each dataset.
- Missing values are identified and summarised.
- A visual heatmap highlights patterns of missing data.
- This step ensures that any inconsistencies or incomplete records can be addressed before further analysis.

The decision was made to visiualise the data as after it was identified that there were only missing values in one dataset, a heatmap was created to quickly reveal patterns. This allows the ability for clearer, visual interpretation of the comparison of what was missing due to there being over 7,000 products.
```python
# Step 2: Data Understanding and Quality Check 

# Creating a dictionary to store all data framesdatasets = 
datasets = {
    "Products CSV": products_csv_df,
    "Reviews CSV": reviews_csv_df,
    "Users CSV": users_csv_df,
    "Products JSON": products_json_df,
    "Reviewers JSON": reviewers_json_df
}

# Summary of structure + top missing values
summary = []
for name, df in datasets.items():
    missing = df.isna().sum().sum()
    summary.append([name, df.shape[0], df.shape[1], missing])

summary_df = pd.DataFrame(summary, columns=["Dataset", "Rows", "Columns", "Total Missing"])
display(summary_df)

# Visualise missingness only for the dataset with missing data
plt.figure(figsize=(8,3))
sns.heatmap(products_csv_df.isna(), cbar=False, cmap="coolwarm")
plt.title("Missing Values — Products CSV")
plt.tight_layout()
plt.show()
```

Step 3: Data Cleaning and Preperation After assessing the data structure and quality, the next step is to clean and standardise it to ensure consistency across files before merging.

In this step:

- All collumn names were standardised by converting to lowercase and key fields renamed.
- All missing values were flagged
- DOB was changed into age groups for more effective analysis
- Removed duplicates
- Prepare the data for merging by ensuring join keys were consistant.
- 
The approach to missing data was to flag the data rather than delete the data. This method was most appropriate as it meant that crucial data was not deleted and the record count was preserved for analysis. It meant that the missing data could be filtered out when needed. This allowed for the comparison of products with and without complete information. The deletion would have resulted in 27% of products being removed due to there being over 2,000 missing prices attached.

The decision was made to group raw age values into age groups, this was because the age values were too granular for meaningful segmentation. Furthermore, the retail industry typically targets broad age demographics therefore this is more digestive for the audience of the data. This has also allowed for clearer visualisation and the construction of customer personas. 

```python
# Step 3: Clean & Prepare Data

# Standardise + rename
def clean_cols(df): df.columns = df.columns.str.strip().str.lower(); return df
datasets = [clean_cols(d) for d in [products_csv_df, reviews_csv_df, users_csv_df, products_json_df, reviewers_json_df]]
for d in datasets: d.rename(columns={'uniq_id':'product_id','username':'user_id'}, inplace=True)
products_csv_df, reviews_csv_df, users_csv_df, products_json_df, reviewers_json_df = datasets

# Clean pricing & flag discounts
for c in ['list_price','sale_price']:
    if c in products_json_df: products_json_df[c] = products_json_df[c].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
dfp = products_json_df
dfp['price_missing_flag'] = dfp[['list_price','sale_price']].isna().any(axis=1).astype(int)
dfp['discount_flag'] = ((dfp['sale_price'] < dfp['list_price']) & dfp['sale_price'].notna()).astype(int)
dfp['discount_percent'] = np.where(dfp['discount_flag'], ((dfp['list_price']-dfp['sale_price'])/dfp['list_price']*100).round(2), np.nan)

# Clean user DOB & derive age
if 'dob' in users_csv_df:
    u = users_csv_df
    u['dob'] = pd.to_datetime(u['dob'], format='%d.%m.%Y', errors='coerce')
    u['age'] = ((pd.Timestamp('today') - u['dob']).dt.days/365.25).round(1)
    u['age_group'] = pd.cut(u['age'], [0,25,40,60,100], labels=['<25','25–40','40–60','60+'], right=False)


#  Summary + sample preview
print(f"Missing price: {dfp['price_missing_flag'].sum()} | Discounts: {dfp['discount_flag'].sum()} | Users missing DOB: {u['dob'].isna().sum()}")
display(dfp[['product_id','list_price','sale_price','discount_flag','discount_percent','price_missing_flag']].head())
display(u[['user_id','dob','age','age_group']].head())
```

Step 5: Sentiment Processing 

With the review text cleaned and prepared, the next step was to apply sentiment analysis and combine this enriched review data with product and user information to build customer personas.

In this step:

- The VADER sentiment model was loaded and applied to each review to generate a compound sentiment score.

- Review data was further cleaned by standardising column names, removing empty entries, and validating rating values.

- The cleaned review dataset was merged with product details and user demographic data to create a unified view of each customer’s interactions.


```python
# Clean review data
df = (reviews_csv_df.copy()
      .rename(columns={'review':'review_text','score':'rating'})
      .assign(review_text=lambda x: x['review_text'].astype(str).str.strip())
      .dropna(subset=['review_text']))

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df[df['rating'].between(0, 5)]

# Sentiment scoring
df['sentiment_score'] = df['review_text'].apply(lambda x: sia.polarity_scores(x)['compound'])

print(f"Valid reviews: {len(df):,} | Avg rating: {df['rating'].mean():.2f} | "
      f"Avg sentiment: {df['sentiment_score'].mean():.2f}")
display(df[['user_id','product_id','rating','sentiment_score','review_text']].head(4))

# Merge reviews → products → users
merged = (df
          .merge(products_json_df, on="product_id", how="left")
          .merge(users_csv_df, on="user_id", how="left"))

print(f"Merged dataset: {len(merged):,} rows")
```

Step 6: Data-Driven Persona Segmentation

To create objective and reliable personas, a K-Means clustering model was applied to customers with complete behavioural data.

In this step:

- Only customers with valid values for rating, sentiment, age, review activity and spend were included, ensuring no imputation or artificial data.

- All features were standardised so each metric contributed equally to the clustering process.

- A stable K-Means model (fixed seeds and multiple initialisations) was used to identify four natural behavioural groups.

- Cluster labels were written back to the full dataset, with incomplete cases left unassigned.

- Cluster centroids were converted back to original units (e.g., actual age and spend) for clear interpretation.

```python
# Build user-level persona dataset 
persona_df = (
    merged.groupby("user_id")
    .agg({
        "rating": "mean",                   # average rating per user
        "sentiment_score": "mean",          # average sentiment per user
        "age": "mean",                      # average age
        "product_id": "count",              # number of reviews/products
        "sale_price": "mean",               # average spend
        "category": lambda x: x.mode().iloc[0] if not x.mode().empty else "unknown"  # favourite category
    })
    .rename(columns={
        "rating": "avg_rating",
        "sentiment_score": "avg_sentiment",
        "age": "avg_age",
        "product_id": "review_count",
        "sale_price": "avg_price",
        "category": "fav_category"
    })
    .reset_index()
)

# Select clustering features 
features = ['avg_rating','avg_sentiment','avg_age','review_count','avg_price']

# Keep only customers with complete data (no imputation) 
df_clust = persona_df.dropna(subset=features)
print(f"Using {len(df_clust)} customers for clustering")
print(f"{len(persona_df) - len(df_clust)} customers excluded due to missing values\n")

# Scale features to equalise influence 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clust[features])

# Run stable K-means clustering 
np.random.seed(42)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=50, max_iter=500)
labels = kmeans.fit_predict(X_scaled)

# Store cluster labels back in the full persona_df 
persona_df['cluster'] = pd.NA
persona_df.loc[df_clust.index, 'cluster'] = labels

print("Customers per cluster:")
print(pd.Series(labels).value_counts().sort_index())

# Convert centroids back to original units (for interpretation)
centroids = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=features
).round(2)
centroids.index.name = "cluster"

print("\nCluster centroids (original units):")
display(centroids)
```

Step 7: Persona Comparison Visualisation

This step visualises how the four data-driven personas differ across the key behavioural metrics used in the segmentation.

In this step:

- Final persona labels are applied to each cluster, replacing numeric IDs with meaningful group names.

- Persona-level averages are calculated to capture the typical behaviour of each group.

- These averages are converted into Low / Medium / High bands to place all metrics on a shared scale.

- Minor spacing adjustments prevent overlap while maintaining each metric’s band category.

- A comparative line chart is produced to show how each persona positions across the behavioural dimensions.

This provides a clear visual summary of the distinct characteristics and relative strengths of each persona.

```python
# STEP 1 — Assign final persona labels to each user. The cluster model identifies four behavioural groups. Here,
# each cluster ID is mapped to its corresponding persona label.

persona_labels = {
    0: "Young Positive Actives",
    1: "Premium High-Value Shoppers",
    2: "Loyal Older Enthusiasts",
    3: "Reserved Mid-Age Value Seekers"
}

persona_df["persona_name"] = persona_df["cluster"].map(persona_labels)

# STEP 2 — Define the personas, metrics, and colours used in the comparison visualisation. These represent the key traits
# we want to compare across personas.

personas = list(persona_labels.values())
metrics = ["avg_rating", "avg_sentiment", "avg_age", "review_count", "avg_price"]

colors = {
    "Young Positive Actives": "#82B2FF",
    "Premium High-Value Shoppers": "#F28E8E",
    "Loyal Older Enthusiasts": "#8FD19E",
    "Reserved Mid-Age Value Seekers": "#FFCE67"
}

x = range(len(metrics))

# STEP 3 — Calculate the average value of each behavioural metric for each persona. This provides a single profile
# representing the ‘typical’ customer in each group.

means = (
    persona_df[persona_df["persona_name"].isin(personas)]
    .groupby("persona_name")[metrics]
    .mean()
)
# STEP 4 — Assign meaningful Low / Medium / High bands using real behavioural thresholds.

cat_df = pd.DataFrame(index=means.index)

banding_rules = {
    "avg_rating":     [0, 1.45, 1.52, 5],
    "avg_sentiment":  [0, 0.55, 0.70, 1],
    "avg_age":        [0, 40, 55, 120],
    "review_count":   [0, 6, 8, 50],
    "avg_price":      [0, 150, 300, 9999]   # FIX: all £75–80 personas fall into "Low"
}

for m in metrics:
    bins = banding_rules[m]
    labels = ["Low", "Medium", "High"]
    cat_df[m] = pd.cut(means[m], bins=bins, labels=labels, include_lowest=True)

# Convert categorical bands into numeric positions
band_map = {"Low": 1.0, "Medium": 2.0, "High": 3.0}
band_vals = cat_df.apply(lambda col: col.map(band_map))


for m in metrics:
    vals = means[m]
    for b in [1, 2, 3]:
        mask = band_vals[m] == b
        if mask.sum() > 1:
            scaled = (vals[mask] - vals[mask].min()) / (vals[mask].max() - vals[mask].min() + 1e-12)
            micro_scaled.loc[mask, m] = b + (scaled - 0.5) * 0.25
# STEP 6 — Produce the persona comparison visualisation. This plot shows how each persona positions across the core behavioural dimensions 
# on a unified Low–High scale.
plt.figure(figsize=(14,6))

for persona in personas:
    plt.plot(
        x,
        micro_scaled.loc[persona],
        marker="o",
        markersize=11,
        linewidth=3,
        label=persona,
        color=colors[persona]
    )

plt.yticks([1,2,3], ["Low", "Medium", "High"], fontsize=12)
plt.xticks(x, metrics, rotation=15, fontsize=11)
plt.grid(alpha=0.25, linestyle="--")
plt.title("Persona Comparison (High / Medium / Low)", fontsize=17, weight="bold")
plt.legend(title="Persona", bbox_to_anchor=(1.02,1), loc="upper left")

plt.tight_layout()
plt.show()
```
