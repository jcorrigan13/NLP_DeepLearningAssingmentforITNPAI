```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

rows=5
points=100
sins = np.zeros((rows,points))
for i in range(rows):
    for j in range(points):
        sins[i,j]=i*np.sin(j*(i+1)/20)
```

```python
for s in sins:
    plt.plot(s)
plt.show()
```

# Now use Numpy to perform the following tasks
## Extract the third series and plot it

```python
# Enter your code here
third_series = sins[2]
print(third_series)

plt.plot(sins[2])
plt.show()
```

## Extract the 10th data point from each series and plot them as a series

```python
# Enter your code here
tenth_data_points = sins[: , 9]
print(tenth_data_points)

plt.plot(tenth_data_points, marker='o')
plt.title("10th Data Point from Each Series")
plt.xlabel("Series index")
plt.ylabel("10th data point value")
plt.show()
```

## Plot only those series that go above 2 at any point
- Use list comprehension and the `any` function

```python
# Enter your code here
above_two = [s for s in sins if np.any(s > 2)]

for s in above_two:
    plt.plot(s)

plt.title("Series where any value > 2")
plt.show()
```

## Plot the result of adding all the series together

```python
# Enter your code here

s = sum(sins)
plt.plot(s)
plt.title('All series added together')
plt.show()
```
