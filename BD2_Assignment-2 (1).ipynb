```python

price_pref = merged_all.groupby('username')['price'].mean().reset_index(name='price_preference')
review_count = merged_all.groupby('username').size().reset_index(name='review_count')
category_mode = merged_all.groupby('username')['category'].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
).reset_index(name='category_mode')


user_behavior = (
    price_pref
    .merge(review_count, on='username')
    .merge(category_mode, on='username')
)


persona_df = persona_df.merge(user_behavior, on='username', how='left')


print(persona_df.columns.tolist())
persona_df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select relevant numeric features (fill missing values with 0)
X = persona_df[['avg_rating', 'avg_sentiment', 'age', 'review_count']].fillna(0)

# Scale the data (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
persona_df['cluster'] = kmeans.fit_predict(X_scaled)

# Check the new column
persona_df['cluster'].value_counts()
```

```python
persona_summary = (
    persona_df.groupby('cluster')[['avg_rating', 'avg_sentiment', 'age',
                                   'review_count', 'price_preference']]
    .mean()
    .round(2)
)

display(persona_summary)
```

```python
# Define human-readable persona labels for each cluster
cluster_names = {
    0: "Young Polite Shoppers",
    1: "Middle-Aged Loyalists",
    2: "Engaged Critics",
    3: "Older Premium Buyers"
}

# Map them to a new column
persona_df['persona_label'] = persona_df['cluster'].map(cluster_names)

# Check it worked
persona_df[['username', 'cluster', 'persona_label']].head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Average features by persona
features = ['avg_rating', 'avg_sentiment', 'age', 'review_count', 'price_preference']
persona_means = persona_df.groupby('persona_label')[features].mean()

# Normalise each feature to 0–1 for fair comparison
persona_means = (persona_means - persona_means.min()) / (persona_means.max() - persona_means.min())

# Set up angles for radar chart
categories = list(persona_means.columns)
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # close the circle

# Plot each persona
plt.figure(figsize=(8, 8))
for persona, row in persona_means.iterrows():
    values = row.tolist()
    values += values[:1]
    plt.polar(angles, values, label=persona, linewidth=2)
    plt.fill(angles, values, alpha=0.15)

plt.xticks(angles[:-1], categories, color='grey', size=10)
plt.title("Persona Profiles: Normalised Comparison", size=14, pad=20)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Aggregate metrics per persona
persona_viz = (
    persona_df.groupby('persona_label')
    .agg({
        'avg_sentiment': 'mean',
        'price_preference': 'mean',
        'review_count': 'mean',
        'age': 'mean'
    })
    .reset_index()
)

# Normalise review_count for bubble sizing
persona_viz['size'] = persona_viz['review_count'] * 80  # adjust for scale

# Create the bubble chart
plt.figure(figsize=(10,7))
sns.scatterplot(
    data=persona_viz,
    x='price_preference',
    y='avg_sentiment',
    size='size',
    hue='persona_label',
    palette='coolwarm',
    alpha=0.8,
    legend='full'
)

# Annotate each bubble with the persona name
for i, row in persona_viz.iterrows():
    plt.text(
        row['price_preference'], row['avg_sentiment'] + 0.01, 
        row['persona_label'],
        fontsize=10, weight='bold', ha='center'
    )

# Titles and labels
plt.title("🌌 JC Penney Customer Persona Universe", fontsize=16, pad=20)
plt.xlabel("Average Price Preference ($)", fontsize=12)
plt.ylabel("Average Sentiment (−1 to +1)", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(title="Persona Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare features
features = ['avg_rating', 'avg_sentiment', 'age', 'review_count', 'price_preference']
X = persona_df[features].fillna(0)

# Run PCA for visual compression
pca = PCA(n_components=2)
persona_df[['pca1', 'pca2']] = pca.fit_transform(X)

# Plot the "Persona Orbit"
plt.figure(figsize=(9,7))
sns.scatterplot(
    data=persona_df, x='pca1', y='pca2',
    hue='persona_label', palette='Spectral', s=120, alpha=0.8, edgecolor='black'
)

plt.title("🌠 Persona Orbit Map – Behavioural Distance Between Customer Types", fontsize=14, pad=20)
plt.xlabel("Behavioural Dimension 1")
plt.ylabel("Behavioural Dimension 2")

# Add some annotated cluster centers
centers = persona_df.groupby('persona_label')[['pca1','pca2']].mean()
for label, (x, y) in centers.iterrows():
    plt.text(x, y, label, fontsize=10, weight='bold', ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.4'))

plt.tight_layout()
plt.show()
```
