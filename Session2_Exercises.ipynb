## Processing user input
Write some code to ask the user to enter a number in kilograms and then print the equivalent weight in pounds. Remember that the value they enter will come in as a string. Your program should cope with kilogram weights like 3.5.

```python
kg_input = input("Enter weight in kilograms: ")
kilograms = float(kg_input)
pounds = kilograms * 2.2
print(f"{kilograms} kilograms is equal to {pounds:.2f} pounds.")
```
