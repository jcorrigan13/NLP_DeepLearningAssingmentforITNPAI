Step 3: Data Cleaning and Preperation
After assessing the data structure and quality, the next step is to clean and standardise it to ensure 
consistency across files before merging.

In this step:
- All collumn names were standardised by converting to lowercase and key fields renamed.
- All missing values were flagged
- DOB was changed into age groups for more effective analysis
- Removed duplicates
- Prepare the data for merging by ensuring join keys were consistant


```python
# STEP 3 — Clean and Prepare Data for Analysis
import pandas as pd
import numpy as np
import re

# Standardise column names 
def standardise_columns(df):
    df.columns = df.columns.str.strip().str.lower()
    return df

datasets = [products_csv_df, reviews_csv_df, users_csv_df, products_json_df, reviewers_json_df]
datasets = [standardise_columns(df) for df in datasets]
products_csv_df, reviews_csv_df, users_csv_df, products_json_df, reviewers_json_df = datasets

# Rename key identifier columns for consistency 
rename_map = {'uniq_id': 'product_id', 'username': 'user_id'}
for df in [products_csv_df, reviews_csv_df, users_csv_df, products_json_df, reviewers_json_df]:
    df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns}, inplace=True)

# Clean product pricing information and extract numeric values from list_price and sale_price
for col in ['list_price', 'sale_price']:
    if col in products_json_df.columns:
        products_json_df[col] = (
            products_json_df[col]
            .astype(str)
            .str.extract(r'(\d+\.?\d*)')[0]  # Extract numeric portions like 41.09
            .astype(float)
        )

# Flag missing price values if either list or sale price is missing
products_json_df['price_missing_flag'] = (
    products_json_df[['list_price', 'sale_price']].isna().any(axis=1).astype(int)
)

# Create a discount flag and percentage (only when both prices exist)
if {'list_price', 'sale_price'}.issubset(products_json_df.columns):
    products_json_df['discount_flag'] = (
        (products_json_df['sale_price'] < products_json_df['list_price'])
        & products_json_df['sale_price'].notna()
        & products_json_df['list_price'].notna()
    ).astype(int)

# Discount percentage (only for valid discount cases)
    products_json_df['discount_percent'] = np.where(
        products_json_df['discount_flag'] == 1,
        ((products_json_df['list_price'] - products_json_df['sale_price']) /
         products_json_df['list_price'] * 100).round(2),
        np.nan
    )

# Clean user data
if 'dob' in users_csv_df.columns:
    # Handle UK date format (dd.mm.yyyy)
    users_csv_df['dob'] = pd.to_datetime(users_csv_df['dob'], format='%d.%m.%Y', errors='coerce')

    # Calculate age in years
    users_csv_df['age'] = ((pd.Timestamp('today') - users_csv_df['dob']).dt.days / 365.25).round(1)

    # Create age groups
    bins = [0, 25, 40, 60, 100]
    labels = ['<25', '25–40', '40–60', '60+']
    users_csv_df['age_group'] = pd.cut(users_csv_df['age'], bins=bins, labels=labels, right=False)

# Validate cleaning results 

print(f"Products missing any price: {products_json_df['price_missing_flag'].sum()}")
print(f"Products with discounts: {products_json_df['discount_flag'].sum()}")
print(f"Users with missing DOBs: {users_csv_df['dob'].isna().sum()}")

# --- 6. Preview cleaned data ---
display(products_json_df[['product_id', 'list_price', 'sale_price',
                          'discount_flag', 'discount_percent', 'price_missing_flag']].head(10))

display(users_csv_df[['user_id', 'dob', 'age', 'age_group']].head(10))

    
```

Step 4: Data Analysis and Visualisation

```python
# STEP 4 — Clean and Enhance Reviews Data
!pip install seaborn matplotlib nltk
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
import numpy as np
import pandas as pd

# Make sure the sentiment model is available
nltk.download("vader_lexicon", quiet=True)
sia = SentimentIntensityAnalyzer()

#  Inspect and standardise
print(f"Reviews shape: {reviews_csv_df.shape}")
print(f"Columns: {reviews_csv_df.columns.tolist()}")

reviews_csv_df.columns = reviews_csv_df.columns.str.strip().str.lower()

#  Rename columns for clarity
rename_map = {"review": "review_text", "score": "rating"}
reviews_csv_df.rename(columns={k: v for k, v in rename_map.items() if k in reviews_csv_df.columns}, inplace=True)

#  Clean review text and rating values
reviews_csv_df = reviews_csv_df.dropna(subset=["review_text"])
reviews_csv_df = reviews_csv_df[reviews_csv_df["review_text"].str.strip() != ""]

reviews_csv_df["rating"] = pd.to_numeric(reviews_csv_df["rating"], errors="coerce")
reviews_csv_df = reviews_csv_df[reviews_csv_df["rating"].between(0, 5)]

#  Add sentiment score
print("Calculating sentiment scores...")
reviews_csv_df["sentiment_score"] = reviews_csv_df["review_text"].astype(str).apply(lambda x: sia.polarity_scores(x)["compound"])

#  Summary

print(f"Total valid reviews: {len(reviews_csv_df):,}")
print(f"Average rating: {reviews_csv_df['rating'].mean():.2f}")
print(f"Average sentiment: {reviews_csv_df['sentiment_score'].mean():.2f}")

#  Preview result
display(reviews_csv_df[["user_id", "product_id", "rating", "sentiment_score", "review_text"]].head(10))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
!pip install scikit-learn --quiet

# Merge reviews with products
merged_reviews_products = reviews_csv_df.merge(
    products_json_df,
    on="product_id",
    how="left"
)
print(f"Merged reviews + products: {len(merged_reviews_products):,} rows")

# Merge with users
merged_all = merged_reviews_products.merge(
    users_csv_df,
    on="user_id",
    how="left"
)
print(f"Final merged dataset: {len(merged_all):,} rows")

# Create base persona dataset (aggregate user-level data)
persona_df = (
    merged_all.groupby("user_id")
    .agg({
        "rating": "mean",
        "sentiment_score": "mean",
        "age": "mean",
        "product_id": "count",
        "sale_price": "mean",
        "category": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan
    })
    .reset_index()
)

# Rename columns for clarity
persona_df.rename(columns={
    "rating": "avg_rating",
    "sentiment_score": "avg_sentiment",
    "age": "avg_age",
    "product_id": "review_count",
    "sale_price": "avg_price",
    "category": "fav_category"
}, inplace=True)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select features
features = ['avg_rating','avg_sentiment','avg_age','review_count','avg_price']
X = persona_df[features].fillna(persona_df[features].median())   # safety

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
persona_df['cluster'] = kmeans.fit_predict(X_scaled)

# Show cluster sizes
print("Customers per cluster:")
print(persona_df["cluster"].value_counts())

# Compute cluster centroids
centroids = pd.DataFrame(
    kmeans.cluster_centers_,
    columns=features
)

print("\nCluster centroids:")
display(centroids)

display(persona_df.head(10))

# Check for missing values
print("\nMissing values summary:")
print(persona_df.isna().sum())

# Visualise sentiment vs rating with a regression line
sns.scatterplot(
    data=persona_df,
    x="avg_sentiment",
    y="avg_rating",
    color="steelblue",
    alpha=0.4,
    edgecolor=None
)

sns.regplot(
    data=persona_df,
    x="avg_sentiment",
    y="avg_rating",
    scatter=False,
    color="darkred",
    line_kws={"linewidth": 2}
)

plt.title("Sentiment vs. Rating", fontsize=14, weight="bold")
plt.xlabel("Average Sentiment (-1 to +1)")
plt.ylabel("Average Rating (0–5)")

plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
cluster_map = {
    0: "Young Value Buyers",
    1: "Active Older Shoppers",
    2: "Critical Older Shoppers",
    3: "Premium Buyers"
}

persona_df["persona"] = persona_df["cluster"].map(cluster_map)
```
